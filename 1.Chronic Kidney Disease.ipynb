## Chronic Kidney Disease

```python
import pandas as pd
```

```python
dataSet = pd.read_csv('CKD.csv')
dataSet.head()
```

```python
dataSet.shape
```

```python
dataSet = pd.get_dummies(dataSet, drop_first = True)
dataSet.head()
```

```python
dataSet['classification_yes'].value_counts()
```

```python
dataSet.columns
```

```python
independent = dataSet[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataSet[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size=1/3, random_state=0)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# x_train = sc.fit_transform(x_train)
# x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import BernoulliNB
# from sklearn.naive_bayes import CategoricalNB

```

```python
#params
randomForest_params = {
    'n_estimators':[50, 100],
    'criterion':['gini', 'entropy'],
    'max_features':['auto', 'sqrt', 'log2']
}
decisionTree_params = {
    'criterion':['gini', 'entropy'],
    'splitter' :['best', 'random'],
    'max_features':['auto','sqrt','log2']
}
logiesticRegression_params = {
    'penalty':['l2'],
    'solver':['lbfgs', 'liblinear', 'newton-cg', 'saga'],
    'multi_class':['auto','ovr']
}
KNeighbors_params = {
    'n_neighbors':[5,10,30,40,50],
}
svc_params = {
    'C':[50,100,1000],
    'kernel':['poly','rbf','sigmoid'],
    'gamma':['scale','auto']
}

#navie bayes

gaussianNB_params = {
    'var_smoothing':[2e-11, 2e-10, 1e-9, 1e-10]
}
multinomialNB_params = {
    'alpha':[1.0,0.2,0.3,0.4],
}
complementNB_params = {
    'alpha':[1.0,0.2,0.3,0.4],
    'norm':[True, False]
}
bernoulliNB_params = {
    'alpha':[1.0,0.2,0.3,0.4],
    'fit_prior':[True, False]
}
```

```python
classifierModels = [
#     ('Support Vector Classifier', SVC(), svc_params),
    ('Random Forest Classifier', RandomForestClassifier(), randomForest_params),
    ('Decision Tress classifier', DecisionTreeClassifier(), decisionTree_params),
    ('Logiestic Regression', LogisticRegression(), logiesticRegression_params),
    ('K-Nearest Neighbors', KNeighborsClassifier(), KNeighbors_params),
    ('Gaussian Navie Bayes', GaussianNB(), gaussianNB_params),
    ('Multinomial Navie Bayes', MultinomialNB(), multinomialNB_params),
    ('Complement Navie Bayes', ComplementNB(), complementNB_params),
    ('Bernoulli Navie Bayes', BernoulliNB(), bernoulliNB_params),
]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score

count = 0
bestScore = 0
bestModel = ''

for name, model, params in classifierModels:
    grid = GridSearchCV(model, params, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted',cv=5)
    grid.fit(x_train, y_train)
    count = count + 1
    print(count,":",name)
    print('best params:', grid.best_params_)
    print('best score:', grid.best_score_)
    grid_predications = grid.predict(x_test)
    cm = confusion_matrix(y_test, grid_predications)
    print('confusion of:', name)
    print(cm)
    print('classification report of:', name)
    clf_report = classification_report(y_test,grid_predications)
    print(clf_report)
    print('Area Under the ROC Curve:', name)
    roc = roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
    print(roc)
    if(bestScore < grid.best_score_):
        bestScore = grid.best_score_
        bestModel = name

print('final best model is:',bestModel)
print(bestModel,":",bestScore)
```

## final best model is: Gaussian Navie Bayes
